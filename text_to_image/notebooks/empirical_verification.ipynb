```python
import matplotlib.pyplot as plt
```

```python
cifar_steps_list = [500, 1000, 2000, 3000]
cifar_pearson_list = [0.7386, 0.8162, 0.8794, 0.8652]
cifar_mae_list = [0.820, 0.553, 0.425, 0.445]

celeba_steps_list = [250, 500, 750, 1000]
celeba_pearson_list = [0.6458, 0.7738, 0.9152, 0.9307]
celeba_mae_list = [0.555, 0.412, 0.378, 0.311]

artbench_steps_list = [100, 200, 400, 800]
artbench_pearson_list = [0.4273, 0.6813, 0.7079, 0.7303]
artbench_mae_list = [0.2423, 0.2000, 0.2034, 0.2442]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))
ax1.plot(cifar_steps_list, cifar_pearson_list, marker="o")
ax2.plot(celeba_steps_list, celeba_pearson_list, marker="o")
ax3.plot(artbench_steps_list, artbench_pearson_list, marker="o")

ax1.set_xlabel("Number of Fine-Tuning Steps", size=16)
ax2.set_xlabel("Number of Fine-Tuning Steps", size=16)
ax3.set_xlabel("Number of Fine-Tuning Steps", size=16)

ax1.set_ylabel("Pearson Correlation of Model Properties", size=16)

ax1.set_title("CIFAR-20", size=16)
ax2.set_title("CelebA-HQ", size=16)
ax3.set_title("ArtBench (Post-Impressionism)", size=16)

ax1.tick_params(axis='both', which='major', labelsize=12)
ax2.tick_params(axis='both', which='major', labelsize=12)
ax3.tick_params(axis='both', which='major', labelsize=12)

plt.tight_layout()
plt.show()
fig.savefig("model_behavior_pearson_across_steps.pdf")
```

```python
cifar_step_runtime = 4.37 / 1000
celeba_step_runtime = 11.0 / 500
artbench_step_runtime = 4.5 / 200

cifar_inference_runtime = 14.0
celeba_inference_runtime = 11.9
artbench_inference_runtime = 6.1

cifar_runtime_list = [cifar_step_runtime * steps + cifar_inference_runtime for steps in cifar_steps_list]
celeba_runtime_list = [celeba_step_runtime * steps + celeba_inference_runtime for steps in celeba_steps_list]
artbench_runtime_list = [artbench_step_runtime * steps + artbench_inference_runtime for steps in artbench_steps_list]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))
ax1.plot(cifar_steps_list, cifar_runtime_list, marker="o", color="tab:red")
ax2.plot(celeba_steps_list, celeba_runtime_list, marker="o", color="tab:red")
ax3.plot(artbench_steps_list, artbench_runtime_list, marker="o", color="tab:red")

ax1.set_xlabel("Number of Fine-Tuning Steps", size=16)
ax2.set_xlabel("Number of Fine-Tuning Steps", size=16)
ax3.set_xlabel("Number of Fine-Tuning Steps", size=16)

ax1.set_ylabel("Average Runtime (Minutes)", size=16)

ax1.set_ylim(0, 45)
ax2.set_ylim(0, 45)
ax3.set_ylim(0, 45)

ax1.set_title("CIFAR-20", size=16)
ax2.set_title("CelebA-HQ", size=16)
ax3.set_title("ArtBench (Post-Impressionism)", size=16)

ax1.tick_params(axis='both', which='major', labelsize=12)
ax2.tick_params(axis='both', which='major', labelsize=12)
ax3.tick_params(axis='both', which='major', labelsize=12)

plt.tight_layout()
plt.show()
fig.savefig("runtime_across_steps.pdf")
```
